```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
housing_data = pd.read_csv('datasets/housing.csv')
housing_data.dropna(axis=0, inplace=True)

housing_data.sample(10)
```

```python
household_data = housing_data[['ocean_proximity', 'households']]
household_data.sample(5)
```

```python
grouped_data = household_data.groupby('ocean_proximity', as_index=False).sum()
grouped_data.head()
```

```python
grouped_data.columns
```

```python
plt.pie(grouped_data['households'],
        labels=grouped_data['ocean_proximity'],
        autopct='%.1f'
        )
plt.title('Households by location')
plt.show()
```

```python
COLORS = ['lightblue', 'limegreen', 'orange', 'gold', 'peachpuff']

plt.pie(grouped_data['households'],
        labels=grouped_data['ocean_proximity'],
        autopct='%.1f', colors=COLORS)

plt.title('Households by location')
plt.axis('equal')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,8))

plt.pie(grouped_data['households'],
        labels=grouped_data['ocean_proximity'],
        autopct='%.1f', colors=COLORS)

plt.title('Households by location', loc='left')
plt.axis('equal')
plt.show()
```

```python
explode = (0,0,0,0, 0.2)
fig, ax = plt.subplots(figsize=(8,8))

wedges, texts, autotexts = plt.pie(grouped_data['households'],
        labels=grouped_data['ocean_proximity'],
        autopct='%.1f', colors=COLORS,
        explode=explode)

plt.title('Households by location', loc='left')
plt.axis('equal') # show equal x and y-axis if it's not already the case.
wedges[4].set(edgecolor='blue', linewidth=2)
plt.show()
```
