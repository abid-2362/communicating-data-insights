```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
mpg_data = pd.read_json('datasets/auto-mpg.json')
mpg_data.head()
```

```python
cols = mpg_data.columns[::-1]
mpg_data = mpg_data[cols]
```

```python
mpg_data.head()
```

```python
from numpy import nan

mpg_data.replace('.', nan, inplace=True)
mpg_data.replace('?', nan, inplace=True)

mpg_data = mpg_data.dropna()
mpg_data.head()
```

```python
cols
```

```python
mpg_data_subset = mpg_data[
    ['acceleration', 'horsepower', 'weight', 'mpg']
]

sns.pairplot(mpg_data_subset)
```

```python
sns.pairplot(
    mpg_data,
    vars=['displacement', 'horsepower', 'mpg'],
    hue='cylinders',
    palette='bright'
)
```

```python
corrmat = mpg_data.corr()

fig, ax = plt.subplots(figsize=(12, 8))
sns.heatmap(
    corrmat,
    vmax=0.8,
    square=True, annot=True,
    fmt='.2f', cmap='viridis'
)
```
