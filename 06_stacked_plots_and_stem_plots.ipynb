```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
np_data = pd.read_csv('datasets/national_parks.csv')
np_data.head()
```

```python
np_data.shape
```

```python
x = np_data['Year']
```

```python
y = np.vstack(
    [
        np_data['Badlands'],
        np_data['GrandCanyon'],
        np_data['BryceCanyon'],
    ]
)
y
```

```python
labels = ['Badlands', 'GrandCanyon', 'BryceCanyon']
plt.figure(figsize=(12,8))
plt.stackplot(x, y, labels=labels)
plt.legend(loc='upper left')
plt.show()
```

```python
colors = ['sandybrown', 'tomato', 'skyblue']

plt.figure(figsize=(12,8))
plt.stackplot(x, y, labels=labels, colors=colors, edgecolor='grey')
plt.legend(loc='upper left')
plt.show()
```

```python
np_data.diff()
```

```python
np_data[['Badlands', 'GrandCanyon', 'BryceCanyon']] = np_data[['Badlands', 'GrandCanyon', 'BryceCanyon']].diff()
np_data.loc[np_data['GrandCanyon'] < 0]
```

```python
plt.figure(figsize=(12,8))
plt.stem(np_data['Year'], np_data['GrandCanyon'])

plt.title('Difference in Park visits to GrandCanyon (by year)')
plt.ylabel('x^6')
plt.show()
```

```python
x = np.linspace(0.1, 2 * np.pi, 41)
y = np.sin(x)

plt.stem(x, y)
plt.title('Sine Curve with Stem Plot')
plt.show()
```

```python
x = np.linspace(0.1, 2 * np.pi, 41)
y = np.cos(x)

plt.stem(x, y)
plt.title('Cos Curve with Stem Plot')
plt.show()
```
