```python
import pandas as pd
import seaborn as sns
from numpy import nan
```

```python
mpg_data = pd.read_json('datasets/auto-mpg.json')
mpg_data.replace('.', nan, inplace=True)
mpg_data.replace('?', nan, inplace=True)
mpg_data = mpg_data.dropna()
mpg_data = mpg_data[
    [
        'acceleration', 'car name', 'cylinders', 'displacement', 'horsepower', 'model year',
        'mpg', 'origin', 'weight'
    ]
]

mpg_data.head()
```

```python
sns.lmplot(
    x='horsepower',
    y='mpg',
    data=mpg_data
)
```

```python
sns.lmplot(
    x='horsepower',
    y='mpg',
    data=mpg_data,
    aspect=1.5,
    hue='origin'
)
```

```python
sns.lmplot(
    x='horsepower',
    y='mpg',
    data=mpg_data,
    aspect=1.5,
    hue='origin',
    palette='Set2'
)
```

```python
sns.lmplot(
    x='cylinders',
    y='mpg',
    aspect=1.5,
    data=mpg_data
)
```

```python
sns.lmplot(
    x='weight',
    y='mpg',
    data=mpg_data,
    col='origin',
)
```

```python
sns.pairplot(
    mpg_data,
    height=3,
    vars = ['mpg', 'acceleration', 'horsepower'],
    kind='reg'
)
```
