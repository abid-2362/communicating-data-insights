```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from numpy import nan
```

```python
mpg_data = pd.read_json('datasets/auto-mpg.json')
mpg_data.replace('.', nan, inplace=True)
mpg_data.replace('?', nan, inplace=True)
mpg_data = mpg_data.dropna()
mpg_data = mpg_data[
    [
        'acceleration', 'car name', 'cylinders', 'displacement', 'horsepower', 'model year',
        'mpg', 'origin', 'weight'
    ]
]

mpg_data.head()
```

```python
plt.figure(figsize=(12, 8))
sns.stripplot(x=mpg_data['mpg'])
```

```python
plt.figure(figsize=(12, 8))

sns.stripplot(x='cylinders', y='mpg',
              data=mpg_data, jitter=True)
```

```python
sns.stripplot(x='cylinders', y='mpg',
              data=mpg_data, jitter=True,
              hue='origin', palette='bright')
```

```python
plt.figure(figsize=(30, 8))
sns.swarmplot(x='cylinders', y='mpg',
              data=mpg_data,
              hue='origin', palette='bright',
              dodge=True,
              )
```
