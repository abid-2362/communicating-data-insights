```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
oil_data = pd.read_csv('datasets/MPOB_data1.csv')
oil_data.dropna(axis=0, inplace=True)

oil_data.head()
```

```python
# There are some trailing spaces in column names, lets remove them first
oil_data.columns
```

```python
oil_data.columns = oil_data.columns.str.strip()
oil_data.columns
```

```python
# lets select a subset of data
selected_oil_data = oil_data[['GroundnutOil', 'FishOil']]

# create a boxplot of this selected subset of data
selected_oil_data.boxplot()
```

```python
# Create a boxplot by Year on Lard
oil_data.boxplot(column=['Lard'], by=['Year'], figsize=(12, 8))
```

```python
# Creating a boxplot from numpy arrays instead of pandas DataFrame

# selecting all columns except first 2, i.e. Year and Month
COLUMNS = oil_data.columns[2:].values
COLUMNS
```

```python
# all rows and all columns excluding first 2 i.e. year and month
oil_data_array = oil_data.iloc[:, 2:].values
oil_data_array
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

bp = plt.boxplot(oil_data_array)
plt.xticks(rotation=70)

plt.xticks([1, 2, 3, 4, 5, 6, 7, 8, 9], COLUMNS)
plt.title('Oil Production 2006-2018')

plt.xlabel('Oils')
plt.ylabel('Prices')

plt.show()
```

```python
COLORS = ['maroon', 'limegreen', 'blue',
          'gray', 'gold', 'lightblue',
          'purple', 'maroon', 'aqua']
fig, ax = plt.subplots(figsize=(12, 8))

bp = plt.boxplot(oil_data_array, patch_artist=True, labels=COLUMNS)

for i in range(len(bp['boxes'])):
    bp['boxes'][i].set(facecolor=COLORS[i])
    bp['caps'][2*i].set(color=COLORS[i])

plt.xticks(rotation=70)
plt.title('Oil Production 2006-2018')
plt.xlabel('Oils')
plt.ylabel('Prices')
plt.show()
```

```python
# following properties can be accessed when set patch_artist=True
print('Number of boxes: ', len(bp['boxes']))
print('Number of whiskers: ', len(bp['whiskers']))
print('Number of caps: ', len(bp['caps']))
print('Number of medians: ', len(bp['medians']))
print('Number of fliers: ', len(bp['fliers']))
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

vp = plt.violinplot(oil_data_array)

# plt.xticks(rotation=70)
plt.title('Oil Production 2006-2018')
plt.xlabel('Oils')
plt.ylabel('Prices')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

vp = plt.violinplot(oil_data_array, showmedians=True)

plt.xticks(
    [1,2,3,4,5,6,7,8,9],
    COLUMNS,
    rotation=70
)
plt.title('Oil Production 2006-2018')
plt.xlabel('Oils')
plt.ylabel('Prices')
plt.show()
```

```python
# creating same box plot from DataFrame
selected_oil_data = oil_data.drop(columns=['Year', 'Month'])
selected_oil_data.head()
```
